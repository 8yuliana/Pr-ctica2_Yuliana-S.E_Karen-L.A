```python
#PRÁCTICA 2 _ KAREN LAZO ALARCÓN _ YULIANA SÁNCHEZ ESCOBAR

#Solo documentamos la primera parte del código debido a que las otras dos partes contienen las mismas definiciones
#lo único que cambia es el tiempo de espera

#Pregunta 1. ¿Es suficiente el personal para atender a los pacientes?

import random # para generar la aleatoriedad
import statistics # para las estadisticas

class Persona:
    def __init__(self, tiempo_llegada):
      """
       marca el tiempo de arribo del paciente y lo clasifica en urgente y rutinario
       depende de la aleatoriedad generada
       tomamos 0.2 para el tipo de paciente urgente
        """
      self.tiempo_llegada = tiempo_llegada
      self.tipo_paciente = "Urgente" if random.random() < 0.2 else "Rutinario" # es 20% de probabiliadad por el enunciado

class Cola:
    def __init__(self):
      """ para el orden en que serán atendidas las personas, siguiendo la politica establecida
      """
      self.cola = []

    def agregar_persona(self, persona):
      """ agrega una persona a la cola y la pone al final, significa que tendra que esperar a ser atendido
      """
      self.cola.append(persona)

    def quitar_persona(self):
      """ quita a la primer persona de la cola, lo que significa que esa persona ya sera atendida
      """
      return self.cola.pop(0)

    def esta_vacia(self):
      """ para ver cuando la cola este vacie, es decir, no haya pacientes esperando
      """
      return len(self.cola) == 0

class Simulacion:
    def __init__(self, num_simulaciones, num_personas, tiempo_promedio_servicio):
      """ inicializa los elementos que deseamos usar
      """
      self.num_simulaciones = num_simulaciones # 20 simulaciones
      self.num_personas = num_personas #50,000
      self.tiempo_promedio_servicio = tiempo_promedio_servicio #30 minutos
      self.tiempos_espera_promedio = [] # cuanto tienen que esperar los pacientes a pasar a consulta
      self.tiempos_sistema_promedio = [] # tiempo en que estan los pacientes

    def ejecutar_simulaciones(self):
      """
      para cuando las personas llegan, esperar en ser atendidas, mientras son atentidas y salen
      se haran las simulaciones para ver el comportamiento y las estadisticas para asi determinar los promedios
      """
      for i in range(self.num_simulaciones):
          #que se repita la simulacion la cantidad de veces que queremos
          # usamos listas
            tiempos_espera = []
            tiempos_sistema = []
            for _ in range(self.num_personas):
              # para las 50,000 personas que queremos
                cola = Cola() #los pacientes llegan a formarse en una cola
                reloj = 0
                #para cuando entren y salgan personas del sistema
                tiempo_llegada = 0
                while tiempo_llegada < 1440:  # Simular por 24 horas (1440 minutos) ya que esta abierto todo el tiempo, asi que es el tiempo en que el sistema trabaja
                    tiempo_interarribo = random.uniform(0.5, 50) # tiempo en que van llegando las personas
                    tiempo_llegada += tiempo_interarribo
                    if tiempo_llegada > 1440:
                        break # activa una condición externa para salir del bucle antes de que se complete
                    persona = Persona(tiempo_llegada) #representa la llegada de una persona al sistema
                    cola.agregar_persona(persona) #esa persona que llega la pone en la cola
                    if len(cola.cola) == 1:  # Si hay solo una persona en la cola (empieza con la consulta) o si hay mas esperando consulta
                        tiempo_servicio = random.expovariate(1 / self.tiempo_promedio_servicio) # Se genera un tiempo aleatorio de servicio para la persona que está siendo atendida, utilizando una distribución exponencial (porque son eventos aleatorios e indep).
                        reloj = max(tiempo_llegada, reloj) + tiempo_servicio # para las estadisticas
                        tiempos_sistema.append(tiempo_servicio)
                    else:
                        tiempo_servicio = random.expovariate(1 / self.tiempo_promedio_servicio)
                        reloj += tiempo_servicio
                        tiempos_sistema.append(tiempo_servicio)
                    persona_atendida = cola.quitar_persona()
                    tiempo_espera = reloj - persona_atendida.tiempo_llegada
                    tiempos_espera.append(tiempo_espera)
                    #los promedios de tiempo espera y tiempo sistema se aregan a las listas
                    #statistics.mean es para obtener el promedio
            self.tiempos_espera_promedio.append(statistics.mean(tiempos_espera))
            self.tiempos_sistema_promedio.append(statistics.mean(tiempos_sistema))
            print(f"Resultados de la simulación {i + 1}: Tiempo promedio de espera: {statistics.mean(tiempos_espera):.2f} minutos, Tiempo promedio en el sistema: {statistics.mean(tiempos_sistema):.2f} minutos")

    def generar_reporte(self):
        print("Reporte de estadísticas después de {} simulaciones:".format(self.num_simulaciones))
        print("Tiempo promedio de espera de los pacientes: {:.2f} minutos".format(statistics.mean(self.tiempos_espera_promedio)))
        print("Tiempo promedio en el sistema: {:.2f} minutos".format(statistics.mean(self.tiempos_sistema_promedio)))

# Realizar 20 simulaciones
simulacion = Simulacion(num_simulaciones=20, num_personas=50000, tiempo_promedio_servicio=30)
simulacion.ejecutar_simulaciones()
simulacion.generar_reporte()
```

```python
#pregunta 3

import random
import statistics

class Persona:
    def __init__(self, tiempo_llegada):
        self.tiempo_llegada = tiempo_llegada
        self.tipo_paciente = "Urgente" if random.random() < 0.2 else "Rutinario"

class Cola:
    def __init__(self):
        self.cola_urgentes = []  # Cola para pacientes urgentes
        self.cola_rutinarios = []  # Cola para pacientes rutinarios

    def agregar_persona(self, persona):
        if persona.tipo_paciente == "Urgente":
            self.cola_urgentes.append(persona)
        else:
            self.cola_rutinarios.append(persona)

    def quitar_persona(self):
        if self.cola_urgentes:
            return self.cola_urgentes.pop(0)  # Atender primero a urgente
        elif self.cola_rutinarios:
            return self.cola_rutinarios.pop(0)  # Si no hay urgente, atender a rutinario
        else:
            return None

    def esta_vacia(self):
        return len(self.cola_urgentes) == 0 and len(self.cola_rutinarios) == 0

class Simulacion:
    def __init__(self, num_simulaciones, num_personas, tiempo_promedio_servicio):
        self.num_simulaciones = num_simulaciones
        self.num_personas = num_personas
        self.tiempo_promedio_servicio = tiempo_promedio_servicio
        self.tiempos_espera_urgentes = []
        self.tiempos_sistema_urgentes = []
        self.tiempos_espera_rutinarios = []
        self.tiempos_sistema_rutinarios = []

    def ejecutar_simulaciones(self):
        for i in range(self.num_simulaciones):
            tiempos_espera_urgentes = []
            tiempos_sistema_urgentes = []
            tiempos_espera_rutinarios = []
            tiempos_sistema_rutinarios = []
            for _ in range(self.num_personas):
                cola = Cola()
                reloj = 0
                tiempo_llegada = 0
                while tiempo_llegada < 1440:  # Simular por 24 horas (1440 minutos)
                    tiempo_interarribo = random.uniform(0.5, 50)
                    tiempo_llegada += tiempo_interarribo
                    if tiempo_llegada > 1440:
                        break
                    persona = Persona(tiempo_llegada)
                    cola.agregar_persona(persona)
                    if len(cola.cola_urgentes) + len(cola.cola_rutinarios) == 1:  # Si es el primer paciente en la cola
                        tiempo_servicio = self.tiempo_promedio_servicio
                        reloj = max(tiempo_llegada, reloj) + tiempo_servicio
                        if persona.tipo_paciente == "Urgente":
                            tiempos_sistema_urgentes.append(tiempo_servicio)
                        else:
                            tiempos_sistema_rutinarios.append(tiempo_servicio)
                    else:
                        persona_atendida = cola.quitar_persona()
                        if persona_atendida:
                            tiempo_servicio = self.tiempo_promedio_servicio
                            reloj += tiempo_servicio
                            tiempo_espera = max(0, reloj - persona_atendida.tiempo_llegada)  # Tiempo de espera nunca puede ser negativo
                            if persona_atendida.tipo_paciente == "Urgente":
                                tiempos_espera_urgentes.append(tiempo_espera)
                                tiempos_sistema_urgentes.append(tiempo_servicio)
                            else:
                                tiempos_espera_rutinarios.append(tiempo_espera)
                                tiempos_sistema_rutinarios.append(tiempo_servicio)
            self.tiempos_espera_urgentes.append(statistics.mean(tiempos_espera_urgentes))
            self.tiempos_sistema_urgentes.append(statistics.mean(tiempos_sistema_urgentes))
            self.tiempos_espera_rutinarios.append(statistics.mean(tiempos_espera_rutinarios))
            self.tiempos_sistema_rutinarios.append(statistics.mean(tiempos_sistema_rutinarios))

    def generar_reporte(self):
        print("Reporte de estadísticas después de {} simulaciones:".format(self.num_simulaciones))
        print("Tiempo promedio de espera de los pacientes urgentes: {:.2f} minutos".format(statistics.mean(self.tiempos_espera_urgentes)))
        print("Tiempo promedio en el sistema de los pacientes urgentes: {:.2f} minutos".format(statistics.mean(self.tiempos_sistema_urgentes)))
        print("Tiempo promedio de espera de los pacientes rutinarios: {:.2f} minutos".format(statistics.mean(self.tiempos_espera_rutinarios)))
        print("Tiempo promedio en el sistema de los pacientes rutinarios: {:.2f} minutos".format(statistics.mean(self.tiempos_sistema_rutinarios)))

# Realizar 20 simulaciones con 50,000 personas y tiempo de atención promedio de 20 minutos
simulacion = Simulacion(num_simulaciones=20, num_personas=50000, tiempo_promedio_servicio=20)
simulacion.ejecutar_simulaciones()
simulacion.generar_reporte()
```
